```python
#3 Demonstrate logical operators (and, or, not) with two conditions.
a=int(input("Enter first number: "))
b=int(input("Enter second number: "))
result1= a>50 and b<60
print(result1)
result2= a>50 or b<60
print(result2)
result3= not(a>50)
print(result3)
```
